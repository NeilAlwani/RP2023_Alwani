```python
import matplotlib.pyplot as plt

# Given data
lat_dim = [3, 5, 7, 9, 11, 13, 15]
mse_ppca = [0.2177, 0.1754, 0.1366, 0.0974, 0.0598, 0.0291, 0.0131]
mse_vae = [0.0586, 0.0591, 0.0586, 0.0590, 0.0594, 0.0589, 0.0593]
mse_cvae = [0.0545, 0.0541, 0.0548, 0.0548, 0.0549, 0.0550, 0.0548]

plt.plot(lat_dim, mse_ppca, marker='o', linestyle='-', color='g', label='ppca')
plt.plot(lat_dim, mse_vae, marker='o', linestyle='-', color='r', label='vae')
plt.plot(lat_dim, mse_cvae, marker='o', linestyle='-', color='b', label='cvae')

# Set y-axis to logarithmic scale
plt.yscale('log')

# Add labels and title
plt.xlabel('Latent Dimension')
plt.ylabel('MSE')
plt.title('Trendline for MSE vs. Latent Dimension')

# Show the plot
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import numpy as np

lat_dim = [3, 5, 7, 9, 11, 13, 15]
kl_ppca = np.array([2.0153, 1.8956, 1.7468, 1.5181, 1.0991, 0.5015, 0.2201])
kl_vae = np.array([0.0013, 0.0008, 0.0013, 0.0007, 0.0006, 0.0009, 0.0006])
kl_cvae = np.array([0.0009, 0.0010, 0.0004, 0.0005, 0.0003, 0.0003, 0.0003])

# Create a custom formatter function
def format_ticks(value, pos):
    return f"{value:.4f}"

# Plotting with logarithmic scale for y-axis
plt.plot(lat_dim, kl_ppca, marker='o', linestyle='-', color='g', label='ppca')
plt.plot(lat_dim, kl_vae, marker='o', linestyle='-', color='r', label='vae')
plt.plot(lat_dim, kl_cvae, marker='o', linestyle='-', color='b', label='cvae')

# Set the custom formatter for y-axis
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_ticks))

# Set logarithmic scale for y-axis
plt.yscale('log')

# Add labels and title
plt.xlabel('Latent Dimension')
plt.ylabel('KL-div (log scale)')
plt.title('Trendline for KL-div vs. Latent Dimension')

# Show the plot
plt.legend()
plt.show()

```
